```python
a=list(cities.location)
import folium
d=[]
for i in a:
    b=(i.split(','))
    d.append(list(map(float,b)))
d[:5]
```

```python
ko=[]
for i in ci[:200]:
    try:
        apikey = '817ce650c352236acc38bb69c1cd8b4e'
        city=i
        baseUrl = 'http://api.openweathermap.org/data/2.5/weather'
        url=f'{baseUrl}?q={city}&appid={apikey}'
        response = requests.get(url)
        du=response.json()
        ko.append(du['main']['temp_min']-273.15)
    except:
        ko.append('NaN')
        pass
len(ko)
```

```python
m=folium.Map(location=d[474],zoom_start=5)

for i in range(len(d)):
    if i<200:
        folium.Marker(d[i],popup=(f'{ci[i]} {ko[i]}'),icon=folium.Icon(icon='cloud')).add_to(m)
    #else:
        #folium.Marker(d[i],popup=('NAN')).add_to(m)
```
